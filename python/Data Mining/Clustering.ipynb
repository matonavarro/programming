## Clustering - Hierarchy

# IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import os
%matplotlib inline
```

## Data  
### US State data regarding a mix of categories such as common web searches, 5 personality dimensions, regional and psychiatric divisons

# LOAD DATA
## Read CSV

```python
dirpath = os.getcwd()
states = pd.read_csv(dirpath+'\\ClusterData.csv')
list(states.columns.values)
```

## Save numerical data only for cluster analysis

```python
st = states[states.columns[2:]]
st.index = states.ix[:,1]
```

# CLUSTERING
## Create Linkage Matrix for clustering  
### How far each point is from every other point

```python
Z = linkage(st, 'ward')
```

## Plot Dendrogram of Clusters  
### A dendogram is a diagram that shows the hierchical relationship between objects.  Used to work out the best way to allocate objects to clusters.  Does not tell you how many clusters you should have

```python
plt.figure(figsize = (25, 10))
plt.title('Cluster with All Searches and Personality')
plt.ylabel('distance')
dendrogram(
    Z,
    labels = st.index,
    leaf_rotation = 0.,
    leaf_font_size = 18.,
)
plt.show()
```

### Heights reflect the distance between clusters.  AL and MI are the most similar states in terms of the variables in used in the data set.
